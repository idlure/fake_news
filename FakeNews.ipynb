```python
#import the libararies
from bs4 import BeautifulSoup
import pandas as pd
import requests
import urllib.request
import time
```

```python
#Create lists to store the scrapped data
quote_statements = []
paragraph = []
```

```python
#Create a function to scrape the website
def scrape_website(page_number):
  page_num = str(page_number)
  URL = 'https://www.ndtv.com/india/page-'+page_num
  webpage = requests.get(URL)
  soup = BeautifulSoup(webpage.text, 'html.parser')

#Get the location of the information
  statement_quote = soup.find_all('div',attrs = {'class': 'news_Itm-cont'}) #Location of the statement

#Loop through the statement_quote
  for i in statement_quote:
    link2 = i.find_all('a')
    link3 = i.find_all('p')
    statement_text_quote = link2[0].text.strip()
    statement_text_paragraph = link3[0].text.strip()
    quote_statements.append(statement_text_quote)
    paragraph.append(statement_text_paragraph)
```

```python
#Loop through n to 'm-1' webpages to scrape the data
n,m = 1,4
for i in range(n,m):
  scrape_website(i)
```

```python
#Create the dataframe
data = pd.DataFrame(columns = ['quotes','paragraph'])
data['quotes'] = quote_statements
data['paragraph'] = paragraph

#show the data
data
```
